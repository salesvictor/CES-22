```python
import turtle

def make_window(title, color = 'white'):
    """
    Set up the window with the given background color and title.
    Returns the new window.
    """
    
    w = turtle.Screen()
    w.bgcolor(color)
    w.title(title)
    
    return w


def make_turtle(color='black', size=1, shape='classic', visible=True):
    """
    Set up a turtle with the given color and pensize.
    Returns the new turtle.
    """
    
    t = turtle.Turtle(visible=visible)
    t.color(color)
    t.shape(shape)
    t.pensize(size)
    
    return t


def draw_pentagram(trt=None, pos=None):
    """
    Draws a pentagram.
    
    If trt is None then a new turtle is created for the drawing.
    Else, the specified turtle is used.
    """
    
    if trt is None:
        trt = make_turtle(visible=False)
        #trt.hideturtle()

    if pos is not None:
        trt.penup()
        trt.goto(pos)
        trt.pendown()
    
    trt.left(36)
    for i in range(5):
        trt.forward(80)
        trt.left(144)

wn = make_window('Exercise 4.10: A pentagram of pentagrams')
speedy = make_turtle(visible=False)
speedy.hideturtle()
speedy.penup()

for i in range(5):
    speedy.forward(120)
    speedy.left(72)
    draw_pentagram(pos=speedy.pos())

wn.mainloop()
```
